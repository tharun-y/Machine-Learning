```python
import pandas as pd 
df = pd.read_csv('Datasets/electricVehicles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Handling Missing Data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df['model'] = imputer.fit_transform(df[['model']]).ravel()
```

```python
df['number_of_cells'] = imputer.fit_transform(df[['number_of_cells']]).ravel()
```

```python
df['torque_nm'] = imputer.fit_transform(df[['torque_nm']]).ravel()
```

```python
df['cargo_volume_l'] = imputer.fit_transform(df[['cargo_volume_l']]).ravel()
```

```python
df['towing_capacity_kg'] = imputer.fit_transform(df[['towing_capacity_kg']]).ravel()
```

```python
df['fast_charge_port'] = imputer.fit_transform(df[['fast_charge_port']]).ravel()
```

```python
df['fast_charging_power_kw_dc'] = imputer.fit_transform(df[['fast_charging_power_kw_dc']]).ravel()
```

```python
df.isnull().sum()
```

```python
df
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
df['top_speed_kmh'] = scaler.fit_transform(df[['top_speed_kmh']])
```

```python
df['battery_capacity_kWh'] = scaler.fit_transform(df[['battery_capacity_kWh']])
```

```python
df['number_of_cells'] = scaler.fit_transform(df[['number_of_cells']])
```

```python
new_df = df.select_dtypes(include=['number']).columns
df[new_df] = scaler.fit_transform(df[new_df])
```

```python
df
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
df['drivetrain'] = le.fit_transform(df['drivetrain'])
```

```python
df['segment'] = le.fit_transform(df['segment'])
```

```python
df['car_body_type'] = le.fit_transform(df['car_body_type'])
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories = [['Lead Acid' , 'Metal Hydride' , 'Lithium-ion']])
df['battery_type'] = encoder.fit_transform(df[['battery_type']])
```

```python
df['fast_charge_port'] = le.fit_transform(df['fast_charge_port'])
```
