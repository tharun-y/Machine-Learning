```python
import pandas as pd
df = pd.read_csv('../../Datasets/spotify_history.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Handling Missing Data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df['reason_end'] = imputer.fit_transform(df[['reason_end']]).ravel()
df['reason_start'] = imputer.fit_transform(df[['reason_start']]).ravel()
```

```python
df.isnull().sum()
```

```python
df['platform'].unique()
```

```python
# Feature Scaling 
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
df['scaled_ms_played']= mms.fit_transform(df[['ms_played']]).ravel()
```

```python
# Encoding Categorical Data 
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df['platform_encoded'] = le.fit_transform(df['platform']).ravel()
df['reason_start_encoded'] = le.fit_transform(df['reason_start']).ravel()
df['reason_end_encoded'] = le.fit_transform(df['reason_end']).ravel()
df['shuffle_encoded'] = le.fit_transform(df['shuffle']).ravel()
df['skipped_encoded'] = le.fit_transform(df['skipped']).ravel()
```

```python
df
```

```python
# Dropping unnecessary columns 
df = df.drop(columns = ['spotify_track_uri','ts' , 'track_name' , 'artist_name' , 'album_name'])
```

```python
df
```

```python
# Separate Features and Target 
X = df.drop(columns = ['ms_played'])
y = df['ms_played']
```

```python
# Splitting Data into Training and the Test sets 
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.2 , random_state= 1)
```
