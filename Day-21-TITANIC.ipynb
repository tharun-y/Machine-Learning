```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')  
df.head()
```

```python
df.isnull().sum()
```

```python
# Handling Missing Values
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(strategy='mean')
df['age'] = imputer.fit_transform(df[['age']])
imputer2 = SimpleImputer(strategy='most_frequent')
df['deck'] = imputer2.fit_transform(df[['deck']]).ravel()
df['embark_town'] = imputer2.fit_transform(df[['embark_town']]).ravel()
df['embarked'] = imputer2.fit_transform(df[['embarked']]).ravel()
```

```python
df.isnull().sum()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['fare'] = scaler.fit_transform(df[['fare']])
```

```python
df.head()
```

```python
# Encoding Categorical
from sklearn.preprocessing import LabelEncoder , OrdinalEncoder
encoder = LabelEncoder()
encoder2 = OrdinalEncoder(categories = [['First' , 'Second' , 'Third']])
df['sex_encoded'] = encoder.fit_transform(df[['sex']])
df['class_encoded'] = encoder2.fit_transform(df[['class']])
df['who_encoded'] = encoder.fit_transform(df[['who']])
df['adult_male_encoded'] = encoder.fit_transform(df[['adult_male']])
df['embark_town_encoded'] = encoder.fit_transform(df[['embark_town']])
df['alive_encoded'] = encoder.fit_transform(df[['alive']])
df['alone_encoded'] = encoder.fit_transform(df[['alone']])
```

```python
df
```

```python
custom_bins = [0,30,50,70,100]
custom_labels = ['Youth' , 'YoungAdult' , 'MiddleAged' , 'Senior']
df['age_bins'] = pd.cut(df['age'] , bins = custom_bins , labels = custom_labels)
```
