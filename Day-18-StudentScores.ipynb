```python
import pandas as pd
df = pd.read_csv('../../Datasets/student_scores.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Spiltting the Independent and Dependent Variables 
X = df.iloc[: , : -1] 
y = df.iloc[: , -1]
```

```python
# Splitting into the Training and Test sets
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2,random_state=0)
```

```python
# Training the Data 
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Visualising Training set
import seaborn as sns
df_train = pd.DataFrame({'X' : X_train.values.flatten() , 'yTrain' : y_train , 'y_pred' :regressor.predict(X_train)})
sns.scatterplot(x='X' , y ='yTrain' , data=df_train , label ='Actual' , color = 'red')
sns.lineplot(x='X' , y ='y_pred' , data = df_train , label='Predicted' , color = 'blue')
```

```python
# Visualising Test set 
df_test = pd.DataFrame({'X' : X_test.values.flatten() , 'y' : y_test , 'y_pred' : regressor.predict(X_test)})
sns.scatterplot(x='X' , y='y' , data = df_test , color = 'red' , label ='Actual')
sns.lineplot(x='X' , y='y_pred' , data= df_test , color='blue' , label='Predicted')
```
